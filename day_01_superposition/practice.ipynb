```python
# --- Day 1: Superposition Challenge ---
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt

# 1. Create the circuit
qc = QuantumCircuit(1)

# 2. Visualize starting state |0> (North Pole)
start_state = Statevector(qc)
display(plot_bloch_multivector(start_state, title="Start: Ground State |0>"))

# 3. Apply Hadamard Gate (The "Quantum Coin Flip")
qc.h(0)

# 4. Visualize the result |+> (The Equator)
superposition_state = Statevector(qc)
display(plot_bloch_multivector(superposition_state, title="Result: Superposition |+>"))

# 5. Measure
qc.measure_all()

# 6. Run Simulation (1000 shots)
sim = AerSimulator()
result = sim.run(qc, shots=1000).result()
counts = result.get_counts()

# 7. Show Probabilities
print("\nMeasurement Counts:", counts)
display(plot_histogram(counts))
```
